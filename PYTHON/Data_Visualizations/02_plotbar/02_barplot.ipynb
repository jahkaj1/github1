```python

import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median

kashti=sns.load_dataset("titanic")
kashti

# draw barplot
sns.barplot(x="class", y="fare", hue="sex", data=kashti, color="red", estimator=median,
           palette='Blues')
plt.show()
```

```python
# saturation (Means intensity of color)
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median

kashti=sns.load_dataset("titanic")
kashti

# draw barplot
sns.barplot(x="class", y="fare", hue="sex", data=kashti, color="red", estimator=median, saturation=10,
           palette='Blues')
plt.show()
```

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt

# read a titanic.csv file
# from seaborn library
kashti=sns.load_dataset("titanic")
kashti

# errcolor is 0 is black, 1 is white, 0-1 is grey
# edgecolor is also between 0-1, black to white, 
# edgecolor is called bar edge color
# facecolor is different code mil kar different color bnate hy
# rgba, red, green,blue, opacity of facecolo

sns.barplot(x="class", y="fare", data=kashti,
            linewidth=3, facecolor=(0.3,0.3,0.8,0.1), 
            errcolor="0.3", edgecolor="0.3")
```
